```python
import pandas as pd
pd.set_option("display.max_rows", 100)
pd.set_option('display.max_columns', None)
```

```python
# import the raw dataset

tab = pd.read_csv(r"fifa21 raw data v2.csv")

tab
```

```python
# checking for duplicates

tab[['LongName','Nationality','Age']].loc[tab[['LongName','Nationality','Age']].duplicated()]
```

```python
# verifying the 1st duplicate

tab.loc[tab['LongName'] == 'Adama Traoré'].sort_values('Nationality', ascending=False)
```

```python
# verifying the 2nd duplicate

tab.loc[tab['LongName'] == 'Peng Wang'].sort_values('Nationality', ascending=False)
```

```python
# removing duplicates 

tab = tab.drop([18280,9249])
```

```python
# verifying that the rows were correctly removed
pd.concat(
    [tab.loc[tab['LongName'] == 'Peng Wang'].sort_values('Nationality', ascending=False),
    tab.loc[tab['LongName'] == 'Adama Traoré'].sort_values('Nationality', ascending=False)]
    )
```

```python
# dropping useless columns

tab = tab.drop(columns = ['photoUrl','playerUrl','Name'])
```

```python
# removing unwanted characters from values

tab['Club'] = tab['Club'].str.strip("\n")
tab['W/F'] = tab['W/F'].str.strip("★")
tab['SM'] = tab['SM'].str.strip("★")
tab['IR'] = tab['IR'].str.strip("★")

# replacing '~' with '-'

tab['Contract'] = tab['Contract'].str.replace('~','-')
```

```python
# removing '€' and converting 'M', 'K' to their numeric value

tab['Value'] = tab['Value'].str.strip("€").str.replace(".",'').str.replace('M','000000').str.replace('K','000')
tab['Wage'] = tab['Wage'].str.strip("€").str.replace(".",'').str.replace('M','000000').str.replace('K','000')
tab['Release Clause'] = tab['Release Clause'].str.strip("€").str.replace(".",'').str.replace('M','000000').str.replace('K','000')

# removing unit measures from height and weight

tab['Height'] = tab['Height'].str.replace('[^0-9]','',regex=True)
tab['Weight'] = tab['Weight'].str.replace('[^0-9]','',regex=True)

tab
```

```python
# converting Height and Weight columns to numeric


tab[['Height','Weight','Value','Wage','Release Clause']] = tab[['Height','Weight','Value','Wage','Release Clause']].apply(pd.to_numeric)
tab.info()
```

```python
# renaming headers of the columns

dict = {'Value': 'Value (€)',
        'Wage': 'Wage (€)',
        'Release Clause': 'Release Clause (€)',
        'Height': 'Height (cm)',
        'Weight': 'Weight (kg)',
        '↓OVA': 'OVA',
        'LongName': 'Name'}
 

tab.rename(columns=dict, inplace=True)

tab.head()
```

```python
# empty all the NaN cells

tab = tab.fillna('')
tab = tab.replace('NaN','')

tab
```

```python
tab = tab.sort_values(by ='ID', ascending = 1)
tab = tab.reset_index(drop = True)
tab
```

```python
tab.to_csv(r"fifa21_cleaned.csv", index = False)
```
